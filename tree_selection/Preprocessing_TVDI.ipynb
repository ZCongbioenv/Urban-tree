```python
# transform the images to the same projection
import os
import rasterio
from rasterio.warp import reproject, Resampling, calculate_default_transform

def register_to_reference(src_paths, ref_path, output_folder, dst_crs='EPSG:3006'):
    # Open the reference image to get its spatial parameters
    with rasterio.open(ref_path) as ref:
        ref_transform = ref.transform
        ref_crs = ref.crs
        ref_shape = (ref.height, ref.width)

    # Process each image
    for src_path in src_paths:
        with rasterio.open(src_path) as src:
            transform, width, height = calculate_default_transform(
                src.crs, ref_crs, src.width, src.height, *src.bounds,
                dst_width=ref.width, dst_height=ref.height)

            kwargs = src.meta.copy()
            kwargs.update({
                'crs': ref_crs,
                'transform': ref_transform,
                'width': ref.width,
                'height': ref.height,
                'compress': 'lzw'
            })

            # Setup output path
            output_path = os.path.join(output_folder, f"{os.path.splitext(os.path.basename(src_path))[0]}_corrected.tif")
            
            # Reproject and save the image
            with rasterio.open(output_path, 'w', **kwargs) as dst:
                for i in range(1, src.count + 1):
                    reproject(
                        source=rasterio.band(src, i),
                        destination=rasterio.band(dst, i),
                        src_transform=src.transform,
                        src_crs=src.crs,
                        dst_transform=ref_transform,
                        dst_crs=ref_crs,
                        resampling=Resampling.nearest  # or Resampling.bilinear for continuous data
                    )
            print(f"Reprojected and saved {src_path} to {output_path}")

# Example usage
base_dir = './'
#indices = ['TIRS1', 'SWIR2', 'SWIR1', 'NIR', 'SAVI', 'NDVI', 'MSAVI2', 'NDSI2', 'MSI', 'NDWI', 'GNDVI', 'EVI', 'LST', 'NDMI', 'MNDWI']
#indices = ['NDSI2']
indices = ['EVI2','EVI2_2']
reference_image = './Landsat8_SR_NDVI/NDVI_2015-07-04.tif'  # Adjust path accordingly

for index in indices:
    folder_path = os.path.join(base_dir, f'Landsat8_SR_{index}')
    src_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.tif') and not f.endswith('_corrected.tif')]
    register_to_reference(src_files, reference_image, folder_path)



```

```python
#calculate seasonal values
import os
import numpy as np
import rasterio
from datetime import datetime

def get_season(month):
    """ Determine the meteorological season based on the month. """
    if month in (12, 1, 2):
        return 'winter'
    elif month in (3, 4, 5):
        return 'spring'
    elif month in (6, 7, 8):
        return 'summer'
    elif month in (9, 10, 11):
        return 'fall'

base_dir = './'
#indices = ['TIRS1', 'SWIR2', 'SWIR1', 'NIR', 'SAVI', 'NDVI', 'MSAVI2', 'NDSI2', 'MSI', 'NDWI', 'GNDVI', 'EVI', 'LST', 'NDMI', 'MNDWI']
#indices = ['NDSI2']
indices = ['EVI2_2']#['EVI2','EVI2_2']
for index in indices:
    folder_path = os.path.join(base_dir, f'Landsat8_SR_{index}')
    seasonal_data = {}
    file_paths = {}  # To store file paths for metadata purposes

    for filename in os.listdir(folder_path):
        if filename.endswith('_corrected.tif'):  # Process only corrected images
            date_str = filename.split('_')[2]  # Assuming format is 'Index_Year_Month_Day_corrected.tif'
            date_obj = datetime.strptime(date_str, '%Y-%m-%d')
            year, month = date_obj.year, date_obj.month

            season_year = f"{year-1}_winter" if month != 12 and get_season(month) == 'winter' else f"{year}_{get_season(month)}"

            if season_year not in seasonal_data:
                seasonal_data[season_year] = []
                file_paths[season_year] = filename  # Store one path for metadata

            with rasterio.open(os.path.join(folder_path, filename)) as src:
                data = src.read(1)  # Read the first band
                mask = data == src.nodata  # Mask to identify nodata values
                data[mask] = np.nan  # Convert nodata to NaN
                seasonal_data[season_year].append(data)

    # Calculate and save the mean for each season-year
    for season_year, rasters in seasonal_data.items():
        if rasters:
            stacked_rasters = np.stack(rasters)
            mean_raster = np.nanmean(stacked_rasters, axis=0)  # Calculate mean ignoring NaNs

            with rasterio.open(os.path.join(folder_path, file_paths[season_year])) as src:
                profile = src.profile
                profile.update(dtype=rasterio.float32, count=1, compress='lzw', nodata=src.nodata)

                output_path = os.path.join(folder_path, f"{index}_{season_year}_mean.tif")
                with rasterio.open(output_path, 'w', **profile) as dst:
                    dst.write(mean_raster.astype(rasterio.float32), 1)
            print(f"Saved mean raster: {output_path}")

print("All seasonal means have been calculated and saved.")




```

```python
import os
import numpy as np
import rasterio
import imageio
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize

def create_seasonal_gif_with_correct_order(folder_path, output_filepath):
    """
    Create a GIF from seasonal TIFF files, ensuring proper color representation and correct seasonal and annual order,
    and include a color bar in each frame.
    """
    # Define the order of seasons for sorting
    seasons_order = {'spring': 1, 'summer': 2, 'fall': 3, 'winter': 4}

    # List all files
    all_files = os.listdir(folder_path)
    print(f"All files in {folder_path}: {all_files}")  # Debugging output

    # Filter files by seasonal suffix and ensure they end with a known season mean tag
    files = [f for f in all_files if any(season in f for season in seasons_order) and f.endswith('mean.tif')]
    print(f"Filtered files: {files}")  # Debugging output

    # Sort files by extracting the year and the season from the filename
    files.sort(key=lambda x: (int(x.split('_')[1]), seasons_order[x.split('_')[2]]))

    if not files:
        print(f"No seasonal files found in {folder_path}. Skipping...")
        return

    print(f"Processing {folder_path}...")

    images_for_gif = []
    fig, ax = plt.subplots(figsize=(10, 8))  # Adjusted figure size to better accommodate the color bar
    for file in files:
        file_path = os.path.join(folder_path, file)
        with rasterio.open(file_path) as src:
            data = src.read(1)
            vmin, vmax = np.percentile(data[~np.isnan(data)], (2, 98))  # Percentile clipping for contrast adjustment
            norm = Normalize(vmin=vmin, vmax=vmax)
            img = ax.imshow(data, cmap='viridis', norm=norm)
            # Set title with extracted year and season from the filename
            ax.set_title(f"{file.split('_')[1]} {file.split('_')[2].capitalize()}", fontsize=12)
            plt.axis('off')

        # Handle color bar
        if 'cbar' in locals():
            cbar.update_normal(img)
        else:
            cbar = plt.colorbar(img, ax=ax, orientation='vertical', fraction=0.046, pad=0.04)

        # Draw and convert to image
        fig.canvas.draw()
        image_from_plot = np.frombuffer(fig.canvas.tostring_rgb(), dtype='uint8')
        image_from_plot = image_from_plot.reshape(fig.canvas.get_width_height()[::-1] + (3,))
        images_for_gif.append(image_from_plot)

        ax.clear()  # Clear the axis for the next image, this should ideally remove the old colorbar, but it does not

    # Save as GIF
    imageio.mimsave(output_filepath, images_for_gif, duration=2000, loop=0)
    print(f"GIF saved at {output_filepath}")
    plt.close(fig)

# Base directory and indices
base_dir = './'
indices = ['TIRS1', 'SWIR2', 'SWIR1', 'NIR', 'SAVI', 'NDVI', 'MSAVI2', 'NDSI2', 'MSI', 'NDWI', 'GNDVI', 'EVI', 'LST', 'NDMI', 'MNDWI','TVDI']

for index in indices:
    print(f"Starting processing for {index}")
    folder_path = os.path.join(base_dir, f'Landsat8_SR_{index}')
    output_gif_path = os.path.join(folder_path, f"{index}_seasonal.gif")
    create_seasonal_gif_with_correct_order(folder_path, output_gif_path)

```

```python
# to save the seasonal data as numpy files
import os
import numpy as np
import rasterio

def convert_seasonal_means_to_numpy(folder_path):
    """
    Convert all seasonal mean TIFF images in the specified folder to NumPy .npy files,
    preserving their original shape.

    Parameters:
        folder_path (str): The path to the folder containing seasonal mean TIFF files.
    """
    # Define the suffixes that indicate a file is a seasonal mean image
    seasonal_suffixes = ['_spring_mean.tif', '_summer_mean.tif', '_fall_mean.tif', '_winter_mean.tif']
    
    # List all files that end with a seasonal mean suffix
    tiff_files = [f for f in os.listdir(folder_path) if any(f.endswith(suffix) for suffix in seasonal_suffixes)]
    
    for tiff_file in tiff_files:
        file_path = os.path.join(folder_path, tiff_file)
        with rasterio.open(file_path) as src:
            data = src.read(1)  # Assuming these are single-band images
            # Construct the path for the corresponding NumPy file
            npy_path = os.path.join(folder_path, os.path.splitext(tiff_file)[0] + '.npy')
            np.save(npy_path, data)
            print(f"Converted {tiff_file} to NumPy array and saved as {npy_path}")

# Base directory where the indices folders are located
base_dir = './'
#indices = ['TIRS1', 'SWIR2', 'SWIR1', 'NIR', 'SAVI', 'NDVI', 'MSAVI2', 'NDSI2', 'MSI', 'NDWI', 'GNDVI', 'EVI', 'LST', 'NDMI', 'MNDWI']
#indices = ['NDSI2']
indices = ['EVI2','EVI2_2']
for index in indices:
    print(f"Processing index: {index}")
    folder_path = os.path.join(base_dir, f'Landsat8_SR_{index}')
    convert_seasonal_means_to_numpy(folder_path)

```

```python
# save all in one file
import os
import numpy as np
import rasterio

def aggregate_and_save_seasonal_data_to_numpy(folder_path):
    """
    Aggregate all seasonal mean TIFF images in the specified folder by season and save each season's data 
    across all years into a single NumPy file.

    Parameters:
        folder_path (str): The path to the folder containing seasonal mean TIFF files.
    """
    # Define seasons to process
    seasons = ['spring', 'summer', 'fall', 'winter']
    seasonal_files = {season: [] for season in seasons}  # Dictionary to hold file lists by season

    # Populate the dictionary with files corresponding to each season
    for file in os.listdir(folder_path):
        for season in seasons:
            if f'_{season}_mean.tif' in file:
                seasonal_files[season].append(file)

    # Process each season's files
    for season, files in seasonal_files.items():
        if not files:
            print(f"No files for {season} season in {folder_path}. Skipping...")
            continue
        
        # Stack arrays from each file
        arrays = []
        for file in files:
            file_path = os.path.join(folder_path, file)
            with rasterio.open(file_path) as src:
                data = src.read(1)  # Read the single band data
                #convert zero values to NaN
                data[data == 0.0] = np.nan
                arrays.append(data)

        if arrays:
            # Stack arrays along a new third dimension and calculate the mean across all years
            stacked_arrays = np.stack(arrays, axis=0)
            # You can also compute the mean or any other statistic here if needed:
            # mean_array = np.mean(stacked_arrays, axis=0)

            # Save the stacked array to a NumPy file
            npy_path = os.path.join(folder_path, f'{season}_all_years.npy')
            np.save(npy_path, stacked_arrays)
            print(f"Saved aggregated data for {season} season to {npy_path}")

# Base directory and indices
base_dir = './'
#indices = ['TIRS1', 'SWIR2', 'SWIR1', 'NIR', 'SAVI', 'NDVI', 'MSAVI2', 'NDSI2', 'MSI', 'NDWI', 'GNDVI', 'EVI', 'LST', 'NDMI', 'MNDWI']
#indices = ['NDSI2']
indices = ['EVI2','EVI2_2']
for index in indices:
    print(f"Processing index: {index}")
    folder_path = os.path.join(base_dir, f'Landsat8_SR_{index}')
    aggregate_and_save_seasonal_data_to_numpy(folder_path)
```

```python
# Define NDVI range and step
ndvi_range = np.linspace(ndvi_flat.min(), ndvi_flat.max(), 100)
lst_min = []
lst_max = []

# Calculate min and max LST for each NDVI segment
for ndvi_value in ndvi_range:
    close_indices = np.abs(ndvi_flat - ndvi_value) < 0.01  # find points close to the current NDVI value
    if any(close_indices):
        lst_min.append(np.min(lst_flat[close_indices]))
        lst_max.append(np.max(lst_flat[close_indices]))
    else:
        lst_min.append(np.nan)
        lst_max.append(np.nan)

# Convert lists to arrays for plotting
lst_min = np.array(lst_min)
lst_max = np.array(lst_max)

# Plotting the results with upper and lower bounds
plt.figure(figsize=(10, 6))
plt.scatter(ndvi_flat, lst_flat, color="blue", alpha=0.3, marker='o', label="2015 Data")
plt.plot(ndvi_range, lst_min, 'r-', label="Lower Bound (Wet Edge)")
plt.plot(ndvi_range, lst_max, 'g-', label="Upper Bound (Dry Edge)")
plt.title("NDVI vs LST Scatter Plot with Identified Bounds for 2015")
plt.xlabel("NDVI")
plt.ylabel("LST (K)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load NDVI and LST data for 2015, assuming ndvi_2015 and lst_2015 are loaded as 2D arrays
mask = ~np.isnan(ndvi_2015) & ~np.isnan(lst_2015) & (ndvi_2015 > 0.2) & (lst_2015 > 0)
ndvi_flat = ndvi_2015[mask]
lst_flat = lst_2015[mask]

# Define NDVI bins
ndvi_bins = np.linspace(ndvi_flat.min(), ndvi_flat.max(), num=50)  # Adjust 'num' for finer segmentation
digitized = np.digitize(ndvi_flat, ndvi_bins)  # Assign each NDVI value to a bin

# Initialize lists to hold the boundary values for LST
min_lst_values = []
max_lst_values = []
second_min_lst_values = []
second_max_lst_values = []

# Calculate min and max LST values for each NDVI bin
for i in range(1, len(ndvi_bins)):
    bin_mask = digitized == i
    if any(bin_mask):  # Check if there are any points in this bin
        sorted_lst = np.sort(lst_flat[bin_mask])
        min_lst_values.append(sorted_lst[0])
        max_lst_values.append(sorted_lst[-1])
        
        if len(sorted_lst) > 4:  # Check if there are enough points to consider second min/max
            second_min_lst_values.append(sorted_lst[1])  # Second smallest LST
            second_max_lst_values.append(sorted_lst[-2])  # Second largest LST
        else:
            second_min_lst_values.append(sorted_lst[0])
            second_max_lst_values.append(sorted_lst[-1])

# Corresponding NDVI values for the plot
ndvi_values_for_plot = ndvi_bins[2:]

# Create plot
plt.figure(figsize=(12, 8))
plt.scatter(ndvi_flat, lst_flat, color="lightgray", alpha=0.5, marker='o', label="2015 Data Points")
plt.plot(ndvi_values_for_plot, min_lst_values, 'r-', label="Wet Edge (Min LST)")
plt.plot(ndvi_values_for_plot, max_lst_values, 'b-', label="Dry Edge (Max LST)")
plt.plot(ndvi_values_for_plot, second_min_lst_values, 'r--', label="Second Wet Edge (Second Min LST)")
plt.plot(ndvi_values_for_plot, second_max_lst_values, 'b--', label="Second Dry Edge (Second Max LST)")
plt.title("NDVI vs LST Scatter Plot with Identified Bounds for 2015")
plt.xlabel("NDVI")
plt.ylabel("LST (K)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Convert lists to numpy arrays for regression
X_ndvi_combined_dry = np.concatenate([ndvi_values_for_plot, ndvi_values_for_plot])  # NDVI values for both primary and secondary dry edges
Y_combined_dry = np.concatenate([max_lst_values, second_max_lst_values])  # LST values for both primary and secondary dry edges

X_ndvi_combined_wet = np.concatenate([ndvi_values_for_plot, ndvi_values_for_plot])  # NDVI values for both primary and secondary wet edges
Y_combined_wet = np.concatenate([min_lst_values, second_min_lst_values])  # LST values for both primary and secondary wet edges

# Reshape for sklearn's LinearRegression
X_ndvi_combined_dry = X_ndvi_combined_dry.reshape(-1, 1)
Y_combined_dry = Y_combined_dry.reshape(-1, 1)
X_ndvi_combined_wet = X_ndvi_combined_wet.reshape(-1, 1)
Y_combined_wet = Y_combined_wet.reshape(-1, 1)

# Initialize regression models for combined edges
model_combined_dry = LinearRegression()
model_combined_wet = LinearRegression()

# Fit models to combined edges
model_combined_dry.fit(X_ndvi_combined_dry, Y_combined_dry)
model_combined_wet.fit(X_ndvi_combined_wet, Y_combined_wet)

# Extract coefficients
a_combined_dry = model_combined_dry.coef_[0][0]
b_combined_dry = model_combined_dry.intercept_[0]
a_combined_wet = model_combined_wet.coef_[0][0]
b_combined_wet = model_combined_wet.intercept_[0]

print(f"Combined Dry Edge Equation: LST = {a_combined_dry:.2f} * NDVI + {b_combined_dry:.2f}")
print(f"Combined Wet Edge Equation: LST = {a_combined_wet:.2f} * NDVI + {b_combined_wet:.2f}")

```

```python
# Generate NDVI values for plotting fitted lines
ndvi_line_space = np.linspace(np.min(X_ndvi_combined_dry), np.max(X_ndvi_combined_dry), 500).reshape(-1, 1)

# Calculate LST values using the fitted models
lst_line_combined_dry = model_combined_dry.predict(ndvi_line_space)
lst_line_combined_wet = model_combined_wet.predict(ndvi_line_space)

# Plot
plt.figure(figsize=(12, 8))
plt.scatter(ndvi_flat, lst_flat, color="lightgray", alpha=0.5, marker='o', label="2015 Data Points")
plt.plot(ndvi_line_space, lst_line_combined_dry, 'r-', label="Combined Dry Edge (Fitted)", linewidth=2)
plt.plot(ndvi_line_space, lst_line_combined_wet, 'b-', label="Combined Wet Edge (Fitted)", linewidth=2)
plt.title("NDVI vs LST Scatter Plot with Combined Edge Fits for 2015")
plt.xlabel("NDVI")
plt.ylabel("LST (K)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import os

# Path to the LST and NDVI data for all summer years
summer_lst_all_years = np.load('./Landsat8_SR_LST/summer_all_years.npy')  # shape: [years, 1327, 810], check path
summer_ndvi_all_years = np.load('./Landsat8_SR_NDVI/summer_all_years.npy')  # shape: [years, 1327, 810], check path

# Coefficients from combined regression fits
a_dry = model_combined_dry.coef_[0][0]  # Slope of the combined dry edge
b_dry = model_combined_dry.intercept_[0]  # Intercept of the combined dry edge
a_wet = model_combined_wet.coef_[0][0]  # Slope of the combined wet edge
b_wet = model_combined_wet.intercept_[0]  # Intercept of the combined wet edge

# Output directory for TVDI .npy files
output_dir = "./Landsat8_SR_TVDI_evi2"
os.makedirs(output_dir, exist_ok=True)  # Ensure the directory exists

# Determine the year for the first dataset assuming it's 2013, then increase annually
base_year = 2013

# Process each year's data except for the last one (presumably 2024)
for year_index in range(summer_lst_all_years.shape[0]):  # Excluding the last dataset
    current_year = base_year + year_index
    
    lst = summer_lst_all_years[year_index]
    ndvi = summer_ndvi_all_years[year_index]
    
    # Calculate LST for wet and dry edges using the regression equations
    LSTwet = a_wet * ndvi + b_wet
    LSTdry = a_dry * ndvi + b_dry

    # Calculate TVDI
    denominator = LSTdry - LSTwet
    denominator[denominator == 0] = np.nan  # Avoid division by zero issues
    TVDI = (lst - LSTwet) / denominator
    #TVDI = (LSTdry - lst) / denominator  #SMI
    TVDI = np.clip(TVDI, 0, 1)  # Ensure TVDI values are within the valid range [0, 1]

    # Save the TVDI array to a .npy file, naming it according to the year and descriptor
    output_filepath = os.path.join(output_dir, f"TVDI2_{current_year}_summer_mean.npy")
    np.save(output_filepath, TVDI)
    print(f"TVDI2 for summer {current_year} saved as '{output_filepath}'.")
```
